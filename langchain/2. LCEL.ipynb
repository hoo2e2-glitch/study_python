```python
class NewLineOutputParser(BaseOutputParser):
    def parse(self, text):
        lines = text.split("\n")
        return [line.lstrip("-123456789. ").strip() for line in lines]
```

```python
newline_parser = NewLineOutputParser()
```

```python
newline_parser.parse("""- 1. 햄버거\n- 2. 떡볶이\n- 3. 치킨""")
```

```python
template = ChatPromptTemplate.from_messages({
    ("system", """
        리스트를 생성하는 기계입니다.
        요청한 모든 리스트에 계수는 최대 {max_length}개 까지만 목록으로 표시하세요.
    """),
    ("human", "{question}")
})

prompt = template.format_messages(
    max_length=5,
    question="AI를 잘하려면 어떤거부터 해야함?"
)
```

## 체인(Chain 생성)
- " | " : 파이프 연산자로 체인을 만든다

```python
# list
first_chain = template | chat | NewLineOutputParser() 
```

```python
chain_result = first_chain.invoke({
    "max_length": 5,
    "question": "AI를 잘하려면 어떤거부터 해야함?"
})
```

```python
chain_result
```

```python
# RunnableSequence
# 이전 단계의 출력이 다음 단계의 입력으로 자동 전달되는 파이프라인 객체

print(type(first_chain))
```

## 실습

```python
# 저녁 메뉴 추천(양식, 중식, 한식 선택할 수 있도록)
# 체인을 생성 후 결과를 출력
```

```python
dinner_template = ChatPromptTemplate.from_messages([
    
    ("system", """
        당신은 저녁 메뉴를 추천하는 도우미입니다.
        사람들이 좋아하는 저녁메뉴 종류를 많이 알고 있습니다.
        음식종류는 양식, 중식, 한식이고 설명없이 목록으로 {max_length}개 까지 답변하세요.
    """),
    ("human", """
        {dinner} 저녁메뉴 추천해줘
    """)
])
```

```python
dinner_chain = dinner_template | chat
```

```python
dinner_result = dinner_chain.invoke({
    "max_length": 3,
    "dinner": "양식"
})
```

```python
dinner_result.content
```

## .stream()
- 결과를 실시간 chunk 단위로 쪼개서 응답한다.

```python
dinner_result = dinner_chain.stream({
    "max_length": 3,
    "dinner": "한식"
})
```

```python
for chunk in dinner_result:
    print(chunk.content, end="", flush=True)
```

```python
# ``JSON { }`` / {{}}
template = ChatPromptTemplate.from_messages([
    ("system", """
        넌 주식의 트레이더야
        사용자가 주식과 관련된 질문을 하면 {max_length}개의 주식을 추천해줘
        
        반드시 아래의 JSON 형식으로 답변해줘
        출력 형식:
        {{
            "one": "tesla",
            "two": "samsung",
            ...
            "five": "apple"
        }}
    """),
    ("human", "{question}")
])
```

```python
stock_chain = template | chat
```

```python
stock_result = stock_chain.invoke({
    "max_length": 5,
    "question": "한국인들에게 가장 인기있는 미국주식 추천해줘"
})
```

```python
stock_result.content
```

```python
import json

json.loads(stock_result.content)
```

```python
class JsonOutputParser(BaseOutputParser):
    def parse(self, text):
        return json.loads(text)
```

```python
chat = ChatOpenAI(temperature=0)

stock_chain = template | chat | JsonOutputParser()
```

```python
stock_result = stock_chain.invoke({
    "max_length": 5,
    "question": "한국인들에게 가장 인기있는 미국주식 추천해줘"
})
```
