```python
import time
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

chat = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0
)

prompt = ChatPromptTemplate.from_messages([
    ("system", "당신은 음식 전문가입니다. 짧고 이해하기 쉽게 답변하세요."),
    ("human", "{food}에 대해 알려주세요.")
])

chain = prompt | chat

start_time = time.time()

response = chain.invoke({"food": "김치찌개"})

end_time = time.time()

print(f"경과시간: {end_time - start_time:.2f}초")
print(f"글자수: {len(response.content)}자")
print()
print(response.content)
```
