```python
# 두 수를 더하고 2배 곱한 결과를 출력

def add(num1, num2, callback):
    callback(num1 + num2)

def get_squre(num, callback):
    callback(num * 2)
```

```python
# lambda -> 한줄

add(10, 20, lambda result: get_squre(result, print)) 
```

```python
# 성과 이름을 전달하면 이름을 풀네임으로 "님"자를 붙여서 반환

def get_pull_name(last, first, callback):
    callback(last + first)

def add_nim(full_name, callback)
    callback(full_name + "님")

get_pull_name("홍", "길동", lambda full_name: add_nim(full_name, print))
```

## map, filter

```python
li1 = [(i + 1) *10 for i in range(10)]
```

```python
li1
```

```python
def mul(num):
    return num * 2

# map(함수, 객체)
list(map(mul,li1))

# lambda함수는 값을 리턴
list(map(lambda num: mul(num), li1))
```

```python
# 65 ~ 90
li2 = [i for i in range(65, 91)]

print(li2)
```

```python
li3 = list(map(lambda c: chr(c), li2))
print(li3)
```

```python
# 소문자로 변경

li4 = list(map(lambda c: chr(c + 32), li2))
print(li4)
```

```python
# li2로 aBcDeF... 로 변경 후 li5 만들기

# 65 ~ 90
li2 = [i for i in range(65, 91)]

li5 = []

for i in range(len(li2)):
    if i % 2 == 0:
        li5.append(chr(li2[i]).lower())
    else: li5.append(chr(li2[i]).upper())
    

print(li5)
```

```python
li6 = [i + 1 for i in range(100)]
```

```python
print(li6)
```

```python
# 3의 배수
li7 = list(filter(lambda n: n % 3 == 0, li6))
```

```python
print(li7)
```

```python
# li6에 3의 배수들만 걸러내고 2배로 바꾸기
li8 = list(map(lambda n: n * 2, filter(lambda n: n % 3 == 0, li6)))
```

```python
print(li8)
```

```python
# li4에서 a ~ g까지 filter 걸러내고 li9 만들기

# 소문자로 변경

li4 = list(map(lambda c: chr(c + 32), li2))
print(li4)

li9 = list(filter(lambda c: chr >= 'a' and c < 'h', map(lambda c: chr(c + 32), li2)))
```
