```python
def log_time(original_function):
    def loggin(*args):
        print(datetime.now())
        return original_function(*args)
        # return add_number(10, 20)
    return loggin
```

```python
@log_time
def add_number(num1, num2):
    return num1 + num2
```

```python
add_number(10, 20)
```

```python
def divide(num1, num2):
    try:
        return int(num1) / int(num2)
    except ZeroDivisionError as e:
        print("0으로 나눌 수 없")
    except ValueError as e:
        print("정수만 입력")
    except Exception as e:
        print("알수 없는 오류")
        
```

```python
divide(10, "ㄹ")
```

```python
def subtrect(num1, num2):
    try:
        return num1 - num2
    except TypeError as e:
        print("정수만 계산가능")
```

```python
subtrect("ㄴ", 10)
```

```python
def safe_calc(original_function):
    def exception_warpper(*args):
        try:
            return original_function(*args)
            
        except ZeroDivisionError as e:
            print("0으로 나눌 수 없")
        except ValueError as e:
            print("정수만 입력")
        except Exception as e:
            print("알수 없는 오류")

    return exception_warpper
```

```python
@safe_calc
def divide(num1, num2):
        return num1 / num2
```

```python
divide(10, 0)
```

```python
@safe_calc
def subtrect(num1, num2):
        return num1 - num2
```

```python
# 이름과 정수를 몇 개 입력할 지 모를 때
# "홍길동님이 전달한 총 합: 000"으로 출력해주는 메서드 구현
```

```python
def safe_calc(original_function):
    def exception_warpper(*args, **keywards):
        try:
            return original_function(*args, **keywards)
            
        except ZeroDivisionError as e:
            print("이름을 입력")
        except ValueError as e:
            print("정수만 더할 수 있음")
        except Exception as e:
            print("알수 없는 오류")

    return exception_warpper

@safe_calc
def total_print(*args, **kwards):
    name = kwards["name"]
    total = sum(args)
    result = f"{name}님이 전달한 총 합: {total}"
    print(result)
```

```python
total_print("hong", 10, 20, 30)
```

```python
total_print(10,20,"",30)
```

```python
total_print("hong", 10)
```
