```python
# 타입확인 / 필드구성
class User:
    name: str
    age: int
    address: str
    phone: str

    def __init__(self, name, age, address, phone):
        self.name = name
        self.age = age
        self.address = address
        self.phone = phone

    def Hobby(self, hobby: str) -> None:
        print(f"{self.name}의 취미는 {hobby}입니다.")
```

```python
jang = User("장보고", 20, "서초구", "010-7897-4578")

jang.name
```

```python
# TypedDict: 클래스를 검증하기 위한 하나의 타입의 틀

from typing import TypedDict
```

```python
class User(TypedDict):
    name: str
    address: str
    hobby: str
```

```python
user1_data: User = {
    
    "name": "홍",
    "address": "강남",
    "hobby": "춤"
}
```

```python
User(**user1_data)
```

### pydantic
- 파이썬에서 데이터를 검증하기 위한 라이브러리
- 타입 힌트와 동일하나, 타입이 일치하지 않으면 오류가 발생한다.

```python
from pydantic import BaseModel
```

```python
class User(BaseModel):
    name: str
    address: str
    hobby: str
```

```python
# 키가 더 추가되면 틀 이외에 값은 무시
# 타입이 다를 때 오류 검증
user2_data: User = {
    "name": "장",
    "address": "서초",
    "hobby": "코딩"
}
```

```python
User(**user2_data)
```
