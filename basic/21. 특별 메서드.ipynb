```python
class User:
    name: str
    age: int
    phone: str

    # 생성자
    def __init__(self, name, age, phone):
        self.name = name
        self.age = age
        self.phone = phone

    # toString
    def __str__(self):
        return f"User(name: {self.name},  age: {self.age},  phone: {self.phone})"
```

```python
print(User("fd", 10, "010-1234-5678"))
```

```python
# 회사원
class Employee(BaseModel):
    id: int
    name: str
    imcome: int
```

```python
employee_data1 = {
    "id": 1,
    "name": "이순신",
    "imcome": 20,
}

employee_data2 = {
    "id": 2,
    "name": "장보고",
    "imcome": 20,
}

employee1 = Employee(**employee_data1)
employee2 = Employee(**employee_data2)
```

```python
# 학생
class Student(BaseModel):
    id: int
    name: str
    imcome: int
```

```python
student_data1 = {
    "id": 1,
    "name": "김철수",
    "imcome": 20,
}

student_data2 = {
    "id": 1,
    "name": "김영희",
    "imcome": 20,
}

student1 = Employee(**student_data1)
student2 = Employee(**student_data2)
```

```python
print(student1)
print(student2)
```

```python
print(employee1)
print(employee2)
print(student1)
print(student2)
```

```python
# 객체의 값이 모두 다른 경우 이상이 없음!

print(Employee.__eq__(employee1, employee1))
print(Employee.__eq__(employee1, employee2))
print(Employee.__eq__(employee1, student1))
print(Employee.__eq__(employee1, student2))
```

```python
# 문제 생길수 있는 코드
class Student:
    def __init__(self, id, name):
        self.id = id
        self.name = name
```

```python
student4 = Student(1, "홍길동")
student5 = Student(1, "홍길동")
```

```python
print(Student.__eq__(student4, student5))
print(student4 == student5)
```

```python
# 문제 생길수 있는 코드
class Student:
    def __init__(self, id, name):
        self.id = id
        self.name = name

    def __eq__(self, other):
        
        # 해시주소
        if self.__repr__() == other.__repr__():
            return True

        # 타입
        if isinstance(other, Student):
            return self.id == other.id

        return False
```

```python
student4 = Student(1, "홍길동")
student5 = Student(1, "홍길동")
```

```python
print(Student.__eq__(student4, student5))
print(student4 == student5)
```
