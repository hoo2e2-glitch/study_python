기본 자료형(Data type)

- 숫자 자료형(int)
    - 양의 정수, 음의정수, 0 , 1, 100, ...
- 숫자 자료형(float)
    - 소수점이 있는 수      0.0, 3.14, ...
- 논리 자료형(bool)
    - 참 또는 거짓      True, False, 값이 있을 때
- 문자열 자료형(str)
    - 문자열         '', "", ...

```python
print(100)
```

```python
import sys
```

```python
# sys.getsizeof(): 메모리를 차지하는 바이트 확인
```

```python
# 단위가 byte
# 참조주소

print(sys.getsizeof(100)) // 24 + 4 
print(sys.getsizeof(0.5))
print(sys.getsizeof("홍길동")) // 리스트
print(sys.getsizeof(True))
```

```python
data1 = 0.5
data2 = 10
data3 = "홍길동"
data4 = False
```

```python
# 자동형변환
# 정수 + 정수 = 실수
data1 + data2
```

```python
# 정수 + 문자열 = 오류
data1 + data3
```

```python
# 문자열 + 논리형 = 오류
data3 + data4
```

```python
# 자동형변환
# 논리형 + 정수
print(data4 + data1)
print(True + data1)
```

```python
# 강제 형변환
# 홍길동 + 20
# "홍길동20"
# 타입(바꿀값): 강제형변환

print(int(True))
print(type(int(True)))




print(type("a"))
```

```python
print(str(10) + "홍길동")

# 마지막 줄을 미리보기 시켜줌
```

```python
# 10 -> 10.0

print(type(0.5))

print(float(10))
```

```python
# 실수 -> 정수(소수생략)
print(int(10.5))
```

```python
# 파이썬은 값이 비어있으면 다 False
# js랑 차이점!!

print(bool(""))
print(bool(0))
print(bool([]))
print(bool({}))
```

## 인덱싱(indexing)
- 시퀀스 자료구조에서 인텍스로 값을 접근하는 문법

```python
print(data[3])
```

```python
print(data[-1])
```

## 슬라이싱(slicing)
- 시퀀스 자료구조에서 인텍스 범위부터 인덱스 범위까지 접근하는 문법

```python
data = "이규혁노민균문윤성노규호"
```

```python
# [시작인덱스:종료인덱스]
# 단, 종료인덱스를 포함하지 않는다.

data[0:3]
```

```python
# [시작인덱스:끝까지]
data[-3:]
```

```python
# [처음부터:종료인덱스]
data[:-3]
```

```python
numbers = "123456789"
```

```python
# 홀수만

numbers[0] + numbers[2] + numbers[4] + numbers[6]
```

```python
# 홀수
numbers[::2]
```

```python
# 짝수 가져오고 싶다면
# [::스텝]
numbers[1::2]
```

```python
# numbers 반대로 뒤집기
numbers[::-1]
```

```python
# 슬라이싱 실습:
# 홀수만 가져와서 반대로 뒤집기

numbers[::-2]
numbers[::2][::-1]
```

```python
# 인덱싱과 형변환을 이용해서 -10 만들기

numbers[0] + numbers[-9]
print(str(-10))
```
