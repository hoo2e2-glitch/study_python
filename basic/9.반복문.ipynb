## 컴프리헨션

```python
[i * 2 for i in [10, 20, 30, 40]]
```

```python
{i * 2 for i in [10, 20, 30, 40]}
```

```python
name = '장', '이', '홍'
```

```python
# 빠른포문 인덱스 없음
# 인덱스만들던가
# 일반for문으로 바꾸기

for i in name:
    print(i)
```

```python
# enumerate : 인덱스 만드는 함수
# 인덱스만들던가

# unpacking
for index , name in enumerate(name):
    print(f"{index + 1}번 {name}")
```

```python
# 일반 for문 range()

# range(10) -> ~19
# range(10, 20) -> 10 ~ 19
# range(20, 30) -> 20 ~ 29
# range(10, 20, 2) -> 10 ~ 19 / 홀수

# "a"[0:]
# "a"[10:20]
# "a"[10:20:30]

for i in range(10):
    print(i)
```

```python
# 1 ~ 50까지 홀수만 출력
# 리스트에 담기
# 1 ~ 50까지 반복하여
# 컴프리헨션에 0,1,2,3,4,5만 담아서 출력
```

```python
# 1 ~ 50까지 홀수만 출력
for i in range(1, 51, 2):
    print(i)
    
# 리스트에 담기
set1 = [i * 2 for i in range(1, 51) i <= 5)]
print(set1)

# 1 ~ 50까지 반복하여
# 세트 컴프리헨션에 0,1,2,3,4,5만 담아서 출력
```

## 딕셔너리 빠른 for문

```python
foods = {
    "a": "카레",
    "b": "라면",
    "c": "불고기",
    "d": "떡볶이",
}
```

```python
# index 접근
for i in foods:
    print(foods.get(i))
```

```python
for i in foods.values():
    print(i)
```

```python
for key, food in foods.items():
    print(key, food)
```

## while 문

```python
count = 0

while count < 5:
    print("반복문")
    count += 1
```

```python
# 값을 넣고 비교함 -> (count := count + 1
count = 0

while (count := count + 1) < 5:
    print("반복문")
```
