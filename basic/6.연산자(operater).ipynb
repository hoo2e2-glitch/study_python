```python
# 논리합(|)
# 하나라도 1이라면 -> 1

print(10 | 11)
```

```python
# 논리곱(&)
# 둘 다 1이면 1, 하나라도 0이면 0

print(10 & 11)
```

```python
# xor
# 모두 같으면 0 , 다르면 1
print(10 ^ 11)
```

```python
# not(~)
# -(1 + 10)

print(~10)
```

## 8. 삼항 연산자

```python
bool("True" if 11 == 10 else "False")
```

```python
# 삼항 두 수 비교
num1 = input("첫 번째 정수\n")
num2 = input("두 번째 정수\n")

print(num1 if num1 > num2 else "두 수가 같음" if num1 == num2 else num2)
```

```python
# 삼항연산자 엔터쓰면 안됨
# 들여쓰기로 영역을 구분한다.

print(num1 if num1 > num2 else "두 수가 같음" 
      if num1 == num2 else)
```

```python
# 사용자에게 정수 또는 실수를 입력 받아서
# 정수인지 실수인지 확인 후 삼항 연산자로 결과를 출력하시오.
# ex) 10 -> "정수"
# ex) 11.0 -> "정수"
# ex) 11.3 -> "실수"
```

```python
num1 = input("숫자 입력")
num2 = input("숫자 입력")
```

```python
print(num1)
print(num2)
```

```python
print(num1 if num1 == num2 else "정수" if num1 != num2 else "실수")
```
