```python
# 완전 객체 지향형 언어

# Market
# 마켓

# Customer
# 할인율

class Market:
    def __init__(self, markek_name, product_age, product_price):
        self.markek_name = markek_name
        self.product_age = product_age
        self.product_price = product_price

    def sell(self, customer):
        customer.money -= self.product_price * (100 - customer.discount) / 100

class Customer:
    def __init__(self, name, money, discount):
        self.name = name
        self.money = money
        self.discount = discount
```

```python
cu = Market("CU", "도시락", 5000)
customer = Customer("이", 6000, 10)

cu.sell(customer)
print(customer.money)
```

```python
# Atm, User 객체
# deposit

# Atm.deposit(User, 10000)
# 유저의 돈은 10000 차감,
# 은행에 돈은 10000 입금

# 은행이 가진돈과 유저의 돈 출력
```

```python
class Atm:
    def __init__(self, name):
        self.name = name
        self.users = {}

    def deposit(self, user, money):

        if user.money < money: 
            return "잔액 부족"
        
        if user.name in self.users:
            # 기존의 돈
            self.users.update({user.name: self.users[user.name] + money})
            
        else:
            self.users.update({user.name: money})

        user.money -= money

    def show(self):
        print(self.users)
```

```python
class User:
    def __init__(self, name, money):
        self.name = name
        self.money = money

    def show(self):
        format_money = format(self.money, "_")
        print(f"{self.name}님의 잔 액: {format_money}원")
```

```python
# 클래스 메서드
class Korean:
    # 클래스 변수
    country = "한국"

    # 클래스 메서드
    # 인스턴스X, 클래스변수O
    # @classmethod # 데코레이터
    @classmethod # 데코레이터
    def trip(cls, country):
        if cls.country == country:
            print("국내여행")
        else:
            print("해외여행")    

    # 정적 메서드
    @staticmethod
    def slogan():
        print("Imagine your korea")
```

```python
hong = Korean()

hong.trip("한국")
hong.slogan()

Korean.trip("미국")
Korean.slogan()
```

## 상속(inheritance)

```python
class Person:
    def __init__(self, name):
        self.name = name

# 상속받음 -> ()
class Student(Person):
    def __init__(self, name, age):
        # 부모의 생성자(super)로 호출함
        super().__init__(name)
        self.age = age
```

```python
hong = Person("홍길동")
jang = Student("장보고", 20)

# 부모 -> 타입도 같이 전달
# 모든 자식은 부모타입이다.
# 부모는 자식타입이 아니다.
print(isinstance(hong, Person))
print(isinstance(hong, Student))
print(isinstance(jang, Person))
print(isinstance(jang, Student))
```

```python
# 부모 class
class Animal:
    def __init__(self, name, feed, species):
         self.name = name
         self.feed = feed
         self.species = species

    def feeding(self):
        print(f"{self.name}은 먹이를 먹습니다.")

# 자식 class
class Dog(Animal):
    def __init__(self, name, feed, species):
        super().__init__(name, feed, species)

    def act(self):
        print(f"{self.name}은 산책을 합니다.")
            
class Cat(Animal):
    def __init__(self, name, feed, species):
        super().__init__(name, feed, species)

    def sleep(self):
        print(f"{self.name}은 잠을 자주 잡니다.")
         
```

```python
dog = Dog("누렁이", "잡식", "강아지")
dog.act()
dog.feeding()
```

```python
cat = Cat("야옹이", "츄르", "고양이")
cat.sleep()
cat.feeding()
```

## 다형성

```python
class Car:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def start_engine(self):
        print(f"{self.brand}는 열쇠로 시동을 켭니다")
        
    def stop_engine(self):
        print(f"{self.brand}는 열쇠로 시동을 끕니다")

class SuperCar(Car):
    def start_engine(self):
        print(f"{self.brand}는 음성으로 시동을 켭니다")
    
    def stop_engine(self):
        print(f"{self.brand}는 음성으로 시동을 끕니다")
```

```python
matiz = Car("matiz", 100000)

matiz.start_engine()
matiz.stop_engine()
```

```python
lamborghini = SuperCar("lamborghini", 200000000)
lamborghini.start_engine()
lamborghini.stop_engine()
```

```python
# 휴대폰 필드
# 브랜드(brand), 가격(price)
# 2G 전화하기
# 3G 전화하기, 인터넷
# 4G 전화하기, 인터넷, 영상통화
```

```python
class Phone2G:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def call(self):
        print("call")


class Phone3G(Phone2G):
    def __init__(self, brand, price):
        super().__init__(brand, price)

    def call(self):
        print("call33")

    def internet(self):
        print("internet")


class Phone4G(Phone3G):
    def __init__(self, brand, price):
        super().__init__(brand, price)

    def internet(self):
        print("internet333")

    def videoCall(self):
        print("videoCall")
```

```python
apple = Phone2G("apple", 200000)

apple.call()
```

```python
apple2 = Phone3G("Apple", 2500000)
apple2.call()
apple2.internet()
```

```python
apple3 = Phone4G("apple3", 3300000)
apple3.call()
apple3.internet()
apple3.videoCall()
```
