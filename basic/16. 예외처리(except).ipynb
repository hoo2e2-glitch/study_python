## 커스텀 예외(Custom Exception)

```python
# Exception : 모든 예외의 부모 클래스

class BadWordException(Exception):
    def __init__(self, message):
        super().__init__(message)
```

```python
# 사용자가 "바보"라고 입력하면 커스텀 예외 발생

try:
    expression = input("채팅을 입력하세요")

    if "바보" in expression:
        raise BadWordException("바보 no!")

except BadWordException as e:
    print(e)
```

```python
# 국어, 영어, 수학 점수를 받아서 총점과 평균을 구해주는 프로그램
# 0 ~ 100점 까지만 입력할 수 있습니다.
# 이름을 입력할 수도 있습니다.
# 학번을 입력할 수도 있습니다.
# 평균의 소숫점 자리를 입력받아 처리할 수도 있습니다.
# 예외가 발생한다면 try ~ except 문법으로 해결하세요.
```

```python
class Student:
    def __init__(self, student_id, name, kor, eng, math):
        self.student_id = student_id
        self.name = name
        self.kor = kor
        self.eng = eng
        self.math = math
        self.total = kor + eng + math
        self.aver = self.total / 3

    # 
    def print_score(self, **kwards):

        is_name_print = kwards.get("name", False)
        round_number = kwards.get("round", 0)

        name = f"{self.name}님의 " if is_name_print else ""
        aver = round(self.aver, round_number) if round_number else int(self.aver)
        message = f"{name}총 점수는 : {self.total}점, 평균 : {aver}점"
        print(message)
```

```python
hong = Student(1, "hong", 40, 50, 67)

hong.print_score(name=True, round=3)
```

```python
class ScoreOutOfRangeException(Exception):
    def __init__(self, message):
        super().__init__(message)
```

```python
try:
    kor = int(input("국어 점수를 입력하세요.\nex) 75"))
    math = int(input("수학 점수를 입력하세요.\nex) 75"))
    eng = int(input("영 점수를 입력하세요.\nex) 75"))

    
    if not (0 <= kor <= 100) or not (0 <= math <= 100) or not (0 <= eng <= 100):
        raise ScoreOutOfRangeException("0 ~ 100까지의 점수를 입력하세요.")

    hong = Student(1, "hong", kor, math, eng)

    is_name = eval(input("이름 출력할까요?\nex)True\n"))
    is_round = int(input("소수점 자리 출력값을 입력하세요\nex)0"))
    payload = {}
    
    if is_name:
        payload.update({"name": is_name})

    payload.update({"round": is_round})
    
    hong.print_score(**payload)

except ScoreOutOfRangeException as e:
    print(e)

except ValueError:
    print("타입이 잘못되었습니다.")
```
